### preprocessing the image 
 image pixel belog to 0-255
 divide the data by 255 to scale image value from 0 -1 

```python
xtrain_scl =xtrain/255
xtest_scl = xtest/255
```

```python
xtrain_scl[0]
```

```python
import numpy as np 
np.unique(ytrain)
```

```python
from keras.utils import to_categorical
ytrain_cat = to_categorical(ytrain)
ytest_cat = to_categorical(ytest) 
```

### bulid the cnn (convolutional neural network )

```python
xtrain.shape
```

```python
from keras.models import Sequential
from keras.layers import Convolution2D , MaxPool2D , BatchNormalization , Dropout ,Flatten , Dense
from keras.callbacks import EarlyStopping
```

```python
model = Sequential()

##first convolutional layer
model.add(Convolution2D(filters=16, kernel_size=(3,3), input_shape=(28,28,1)))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(BatchNormalization())
model.add(Dropout(0.2))

## second convolutional layer
model.add(Convolution2D(filters=16,kernel_size=(3,3)))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(BatchNormalization())
model.add(Dropout(0.2))

## flatten layer 
model.add(Flatten())

##dense layer
model.add(Dense(128 , activation='relu'))
model.add(Dense(128, activation='relu'))

## output layer
model.add(Dense(10, activation='softmax')) 
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
early_stop =EarlyStopping(monitor='val_loss',patience=3)
```

```python
hist= model.fit(xtrain_scl, ytrain_cat ,validation_split=0.2,epochs=100,callbacks=[early_stop])
```

## plot the learnig curve 

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(16,8))
plt.plot(hist.history['loss'],label='training loss')
plt.plot(hist.history['val_loss'],label='validation loss')
plt.title('learning curve')
plt.xlabel('epochs')
plt.ylabel('categorical cross entropy loss')
plt.show()
```

### evaluate the model 

```python
model.evaluate(xtrain_scl ,ytrain_cat)
```
