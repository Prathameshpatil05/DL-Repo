### Regression in keras 


```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
from sklearn.datasets import make_regression
x ,y = make_regression(n_samples=1000,n_features=10,n_informative=6,noise=10,random_state=42)
```

```python
x
```

```python
y
```

```python
import pandas as pd
x =pd.DataFrame(x)
x
```

```python
cols = [f'x{i+1}'for i in range (10)]
cols
```

```python
x.columns =cols
x
```

```python
y= pd.DataFrame(y)
y.columns=["y"]
y
```

#### preprocessing on x dataframe - scikit learn pipeline  

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                           ('scaler',StandardScaler())])
```

```python
x_pre =num_pipe.fit_transform(x)
x_pre
```

```python
cols =num_pipe.get_feature_names_out()
cols
```

```python
x_pre=pd.DataFrame(x_pre,columns=cols)
x_pre.head()
```

```python
x_pre.mean()
```

```python
x_pre.std()
```

### train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x_pre, y, test_size=0.2, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

#### create kears model 

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
xtrain.shape
```

```python
## Add intialize the nural network model 
nn=Sequential()

## start adding layer on by one
nn.add(Dense(64, activation='relu', input_dim =xtrain.shape[1]))

## hidden layer 2
nn.add(Dense(32, activation='relu'))

## Output layer 
nn.add(Dense(1, activation='linear'))
```

### complier the model 


```python
nn.compile(optimizer='adam', loss='mse',metrics=['mae'])
```

### fit the nural network model 

```python
Hist = nn.fit(xtrain, ytrain, epochs=100,validation_split=0.2)
```

```python
Hist.history['val_loss']
```

### plot the learning curve

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(Hist.history['loss'],label ='train MSE')
plt.plot(Hist.history['val_loss'], label = 'validation MSE')
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.title('linear curve')
plt.show()
```

##### evalute the model 

```python
nn.evaluate(xtrain, ytrain)
```

```python
nn.evaluate(xtest , ytest)
```

#### predict train and test data 

```python
ypred_tr = nn.predict(xtrain)
ypred_ts =nn.predict(xtest)
```

### R2 score in train and test

```python
from sklearn.metrics import r2_score
tr_r2 =r2_score(ytrain, ypred_tr)
tr_r2
```

```python
ts_r2 = r2_score(ytest , ypred_ts)
ts_r2
```
